# Handling missing data in dataframe
1. fillna to fill massive valuse 
2.interpolate to make a guess on massive values using interpolation
2. dropna to drop rows with massive value

```python
import pandas as pd
```

```python
nw_data = data.fillna(method ='ffill') # carry forward current value
nw_data
```

```python
nw_data = data.fillna(method ='ffill')
nw_data
```

```python
nw_data = data.fillna(method ='ffill', axis ='columns')
nw_data
```

```python
weath =pd.read_csv('wea.csv')
weath
```

```python
nw_waeth = weath.replace([-99999,6.0],np.NaN)
nw_waeth
```
