```python
y = train.iloc[:, -1]
X = train.iloc[:, :-1]
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
train_scaled = scaler.fit_transform(X)
X = pd.DataFrame(train_scaled, columns = X.columns)
X.head()
```

```python
import tensorflow as tf
from tensorflow import  keras
 
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape = (X.shape[1], )),
    tf.keras.layers.Dense(128, activation = "relu"),
    tf.keras.layers.Dense(4, activation=  'softmax')

])
```

```python
weights = model.get_weights()
weights
```

```python
model.summary()
```

```python
model.compile(
    optimizer ='adam',
    loss ="sparse_categorical_crossentropy", 
     metrics= ['accuracy']
     )

history = model.fit(X,y,epochs=50,validation_split=0.2)
```

```python
def he_initialization(shape, dtype =None):
 return tf.random.normal(shape) * np.sqrt(2/ len(shape))
def random_normal_initialization(shape, dtype = None):
  return tf.random.normal(shape)*np.sqrt(1/len(shape))
  
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape = (X.shape[1] )),
    tf.keras.layers.Dense(128, activation = "relu", kernel_initializer = he_initialization),
    tf.keras.layers.Dense(4, activation=  'softmax', kernel_initializer = random_normal_initialization)

])
model.summary()
```

```python
def he_initialization(shape, dtype =None):
 return tf.random.normal(shape) *np.sqrt(2/(len(shape)))
def random_normal_initialization(shape, dtype = None):
  return tf.random.normal(shape)*0.001
  
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape = (X.shape[1] )),
    tf.keras.layers.Dense(128, activation = "relu", kernel_initializer = he_initialization),
    tf.keras.layers.Dense(4, activation=  'softmax', kernel_initializer = random_normal_initialization)

])
model.summary()
```
